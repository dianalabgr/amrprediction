```python
def predict_amr(sample_name, genus_antibiotics):
    results = {'Antibiotic': [], 'Prediction': []}
    
    for antibiotic in genus_antibiotics:
        if antibiotic in models:
            kmer_size = models[antibiotic]['best_kmer']
            kmer_file = f"/home/argis/Desktop/pasteur/amr_prediction_project/amr_ML_blood/BC_test/kmers/kmers/kmer_{kmer_size}_flattened_{sample_name}.fasta_.npz"
            
            try:
                X_npz = np.load(kmer_file, allow_pickle=True)
                X_dna = X_npz['all_kmer_counts_dna'].reshape(1, -1)
                X_rrna = X_npz['all_kmer_counts_rrna'].reshape(1, -1)
                X_protein = X_npz['all_kmer_counts_protein'].reshape(1, -1)
                X = np.concatenate((X_dna, X_rrna, X_protein), axis=1)
                X = csr_matrix(X)
            except FileNotFoundError:
                results['Antibiotic'].append(antibiotic)
                results['Prediction'].append('File Not Found')
                continue
            
            model = models[antibiotic]['model']

            if isinstance(model, xgb.Booster):
                dtest = xgb.DMatrix(X)
                y_pred = model.predict(dtest)
                prediction = "R" if y_pred[0] > 0.5 else "S"
            else:  # XGBClassifier or RandomForest
                y_pred = model.predict(X)
                prediction = "R" if y_pred[0] > 0.5 else "S"

        else:
            prediction = 'NA'
        
        results['Antibiotic'].append(antibiotic)
        results['Prediction'].append(prediction)
    
    return pd.DataFrame(results)
```

```python
import pandas as pd

# Read the CSV file
antibiogram_path = "Petinaki_sample_antibiogram2.csv"  # Update this path
antibiogram_df = pd.read_csv(antibiogram_path)

# Initialize a dictionary to store phenotypes for each sample
sample_phenotypes = {}

# Iterate through each sample (every second column starting from index 1)
for i in range(2, len(antibiogram_df.columns), 2):
    sample = antibiogram_df.columns[i]  # Sample name is in the column header
    sample = sample.split('.')[0]  
    phenotype_col = i  # Phenotype column index
    
    # Initialize a dictionary for the current sample's phenotypes
    phenotypes = {}
    
    # Iterate through each antibiotic (rows)
    for _, row in antibiogram_df.iterrows():
        antibiotic = row[0]  # First column contains antibiotic names
        phenotype = row[phenotype_col]  # Get phenotype for this sample
        
         # Skip if phenotype is NA, IE, or if the key is 'Antibiotic' and value is 'Phe'
        if pd.isna(phenotype) or phenotype == 'IE' or phenotype == 'I' or (antibiotic == 'Antibiotic' and phenotype == 'Phe'):
            continue
        
        # Store the phenotype
        phenotypes[antibiotic] = phenotype
    
    # Add the sample's phenotypes to the main dictionary
    sample_phenotypes[sample] = phenotypes

# Now sample_phenotypes contains cleaned phenotypes for each sample
print(sample_phenotypes)
```

```python
all_antibiotics = set()
for phenotypes in sample_phenotypes.values():
    all_antibiotics.update(phenotypes.keys())
all_antibiotics = sorted(all_antibiotics)  # Sort for consistent ordering

# Initialize a dictionary to store the DataFrame data
data = {'Antibiotic': all_antibiotics}

# Add each sample's phenotypes as a column
for sample, phenotypes in sample_phenotypes.items():
    data[sample] = [phenotypes.get(antibiotic, 'NA') for antibiotic in all_antibiotics]

# Create the DataFrame
antibiogram_df = pd.DataFrame(data)

# Set 'Antibiotic' as the index
antibiogram_df.set_index('Antibiotic', inplace=True)

# Display the DataFrame
print(antibiogram_df)
```

```python
# Read the species file to map samples to their genus
species_path = "petinaki_barcodes_species.txt"  # Update this path
species_df = pd.read_csv(species_path, sep='\t')  # Assuming tab-separated file

# Create a dictionary to map samples to their genus
sample_to_genus = dict(zip(species_df['Sample'], species_df['Species'].str.split().str[0]))

# Define genus-specific antibiotic lists
genus_antibiotics = {
    'Escherichia': escherichia_antibiotics,
    'Klebsiella': klebsiella_antibiotics,
    'Acinetobacter': acinetobacter_antibiotics,
    'Pseudomonas': pseudomonas_antibiotics,
    'Staphylococcus': staphylococcus_antibiotics,
    'Enterobacter': enterobacter_antibiotics,
    'Enterococcus': enterococcus_antibiotics
}

# Initialize a list to store all results
all_results = []

# Process each sample
for sample in sample_phenotypes.keys():
    # Get the genus for the current sample
    genus = sample_to_genus.get(sample, 'Unknown')
    
    # Get the appropriate antibiotic list based on the genus
    if genus in genus_antibiotics:
        antibiotics = genus_antibiotics[genus]
    else:
        print(f"Warning: No antibiotic list found for genus {genus}. Skipping sample {sample}.")
        continue
    
    # Predict the antibiogram for the sample
    results_df = predict_amr(sample, antibiotics)
    
    # Add the sample name as a column
    results_df['Sample'] = sample
    
    # Append the results to the list
    all_results.append(results_df)

# Combine all results into a single DataFrame
final_results_df = pd.concat(all_results, ignore_index=True)

# Display the final results
print(final_results_df)
```

```python
import pandas as pd

# Read Resfinder phenotype table
resfinder_df = pd.read_csv("Resfinder_phenotype.txt", sep="\t", index_col=0)

# Normalize index and columns
resfinder_df.index = resfinder_df.index.str.strip().str.lower()
resfinder_df.columns = resfinder_df.columns.str.strip()

# Normalize final_results_df values
final_results_df['Antibiotic'] = final_results_df['Antibiotic'].str.strip().str.lower()
final_results_df['Sample'] = final_results_df['Sample'].str.strip()

# Add Resfinder phenotype column
def get_resfinder_value(row):
    try:
        return resfinder_df.at[row['Antibiotic'], row['Sample']]
    except KeyError:
        return None

final_results_df['Resfinder'] = final_results_df.apply(get_resfinder_value, axis=1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Initialize a dictionary to store statistics for each antibiotic
antibiotic_stats = {}

# Iterate through each antibiotic
for antibiotic in antibiogram_df.index.unique():
    # Skip if the antibiotic is not in antibiogram_df
    if antibiotic not in antibiogram_df.index:
        antibiotic_stats[antibiotic] = {
            'Accuracy': 'NA',
            'Precision': 'NA',
            'Recall': 'NA',
            'F1-Score': 'NA',
            'TP': 'NA',
            'TN': 'NA',
            'FP': 'NA',
            'FN': 'NA'
        }
        continue
    
    # Filter predictions and actual phenotypes for the current antibiotic
    predictions = final_results_df[(final_results_df['Antibiotic'] == antibiotic) & (final_results_df['Prediction'] != 'NA')]
    predictions = predictions.set_index('Sample')['Prediction']  # Set 'Sample' as the index
    actual = antibiogram_df.loc[antibiotic]
    
    # Align the data (ensure both DataFrames have the same samples)
    common_samples = set(predictions.index).intersection(actual.index)
    predictions = predictions.loc[common_samples]
    actual = actual.loc[common_samples]
    
    # Remove samples where actual antibiogram is NA
    valid_samples = actual[actual != 'NA'].index
    predictions = predictions.loc[valid_samples]
    actual = actual.loc[valid_samples]
    
    # Skip if there are no valid samples
    if len(valid_samples) == 0:
        print(f"No valid samples for antibiotic {antibiotic}. Skipping.")
        antibiotic_stats[antibiotic] = {
            'Accuracy': 'NA',
            'Precision': 'NA',
            'Recall': 'NA',
            'F1-Score': 'NA',
            'TP': 'NA',
            'TN': 'NA',
            'FP': 'NA',
            'FN': 'NA'
        }
        continue
    
    # Calculate statistics
    accuracy = accuracy_score(actual, predictions)
    precision = precision_score(actual, predictions, average='binary', pos_label='R')
    recall = recall_score(actual, predictions, average='binary', pos_label='R')
    f1 = f1_score(actual, predictions, average='binary', pos_label='R')
    
    # Calculate confusion matrix
    cm = confusion_matrix(actual, predictions, labels=['R', 'S'])
    TP, FN, FP, TN = cm.ravel()
    
    # Store statistics and confusion matrix values
    antibiotic_stats[antibiotic] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'TP': TP,
        'TN': TN,
        'FP': FP,
        'FN': FN
    }

# Convert statistics to a DataFrame
stats_df = pd.DataFrame.from_dict(antibiotic_stats, orient='index')

# Display the statistics table
print(stats_df)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd

# Initialize dictionary to store statistics
resfinder_stats = {}

# Iterate through each antibiotic based on antibiogram_df (same as your original loop)
for antibiotic in antibiogram_df.index.unique():
    # Filter Resfinder predictions for this antibiotic
    resfinder_df_filtered = final_results_df[(final_results_df['Antibiotic'] == antibiotic) & (final_results_df['Resfinder'] != 'NA')]
    resfinder_preds = resfinder_df_filtered.set_index('Sample')['Resfinder']

    # Get actual phenotype from antibiogram
    if antibiotic not in antibiogram_df.index:
        resfinder_stats[antibiotic] = {m: 'NA' for m in ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'TP', 'TN', 'FP', 'FN']}
        continue

    actual = antibiogram_df.loc[antibiotic]

    # Align the data (same samples)
    common_samples = set(resfinder_preds.index).intersection(actual.index)
    resfinder_preds = resfinder_preds.loc[common_samples]
    actual = actual.loc[common_samples]

    # Remove samples where actual is 'NA'
    valid_samples = actual[actual != 'NA'].index
    resfinder_preds = resfinder_preds.loc[valid_samples]
    actual = actual.loc[valid_samples]

    # Skip if no valid data
    if len(valid_samples) == 0:
        print(f"No valid samples for antibiotic {antibiotic}. Skipping.")
        resfinder_stats[antibiotic] = {m: 'NA' for m in ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'TP', 'TN', 'FP', 'FN']}
        continue

    # Standardize values
    actual = actual.astype(str).str.strip().str.upper()
    resfinder_preds = resfinder_preds.astype(str).str.strip().str.upper()

    # Keep only R/S
    valid = actual.isin(['R', 'S']) & resfinder_preds.isin(['R', 'S'])
    actual = actual[valid]
    resfinder_preds = resfinder_preds[valid]

    if len(actual) == 0:
        resfinder_stats[antibiotic] = {m: 'NA' for m in ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'TP', 'TN', 'FP', 'FN']}
        continue

    # Metrics
    accuracy = accuracy_score(actual, resfinder_preds)
    precision = precision_score(actual, resfinder_preds, pos_label='R')
    recall = recall_score(actual, resfinder_preds, pos_label='R')
    f1 = f1_score(actual, resfinder_preds, pos_label='R')
    cm = confusion_matrix(actual, resfinder_preds, labels=['R', 'S'])
    TP, FN, FP, TN = cm.ravel() if cm.shape == (2, 2) else (0, 0, 0, 0)

    # Store stats
    resfinder_stats[antibiotic] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'TP': TP,
        'TN': TN,
        'FP': FP,
        'FN': FN
    }

# Convert to DataFrame
resfinder_stats_df = pd.DataFrame.from_dict(resfinder_stats, orient='index')
#print(resfinder_stats_df)
```

```python
"""
In the context of antibiotic resistance prediction, False Positives (FP) represent cases where the model incorrectly predicts resistance when the actual phenotype is sensitive. Here's a breakdown:
Confusion Matrix Terms:
True Positives (TP): Correctly predicted resistance (actual = R, predicted = R).
True Negatives (TN): Correctly predicted sensitivity (actual = S, predicted = S).
False Positives (FP): Incorrectly predicted resistance (actual = S, predicted = R).
False Negatives (FN): Incorrectly predicted sensitivity (actual = R, predicted = S).
Example:
Suppose we have the following predictions and actual phenotypes for an antibiotic:
| Sample | Actual | Predicted |
|--------|--------|-----------|
| BC1 | R | R | (TP)
| BC2 | S | S | (TN)
| BC3 | S | R | (FP)
| BC4 | R | S | (FN)
"""
```
